```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Delete 2 random and add their difference
def play_round(set):
    choice = random.sample(range(len(set)), 2)
    difference = abs(set[choice[0]] - set[choice[1]])
    if choice[0] > choice[1]:
        set.pop(choice[0])
        set.pop(choice[1])
    else:
        set.pop(choice[1])
        set.pop(choice[0])
    set.append(difference)

# play_round until the array is size 1
def play_game(set, size):
    for i in range(size-1):
        play_round(set)
    return set

# Play and plot 10000 games of size 200
# size = 200

# result = np.zeros(size+1)
# for i in range(10000):
#     result[play_game([i for i in range(1, size+1)], size)[0]] += 1

# plt.figure(figsize=(10, 6))
# plt.bar(np.arange(size+1), result)
# plt.title("For 200 numbers")
# plt.xlabel("Final number")
# plt.ylabel("Occurences")
# plt.show()

# Play and plot 10000 games of 2024
size = 2024

result = np.zeros(size+1)
for i in range(20000):
    result[play_game([i for i in range(1, size+1)], size)[0]] += 1

plt.figure(figsize=(10, 6))
plt.bar(np.arange(size+1), result)
plt.title("For 2024 numbers")
plt.xlabel("Final number")
plt.ylabel("Occurences")
plt.show()
```
