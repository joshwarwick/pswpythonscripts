```python
import numpy as np
import matplotlib.pyplot as plt

volume = []
glass_numbers = []
total_volume = []

# Loop through values for case 1:
for i in range(200, 0, -2):
    volume.append(1/i)

    # Find the sum for the number of glasses:
    no_glasses = 0
    for j in range((3*i)//2):
        no_glasses += 3**( (3*i)//2 - j-1)
    glass_numbers.append(no_glasses + 4)
    total_volume.append(1/i * (no_glasses+4))

fig = plt.figure(figsize=(10,6))
a = fig.subplots()
a.set_yscale('log')
a.scatter(volume, glass_numbers)
a.scatter(volume, total_volume)
plt.title('Number of glasses required to win given volume per turn $= 1 / (2 i)$')
plt.xlabel('Volume per turn')
plt.ylabel('log')
a.legend(['No. of glasses', 'Total volume'])
plt.show()
```
