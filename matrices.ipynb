```python
import numpy as np
import itertools
import random

def random_game(size):
    # Generate the array of the permutations
    permutation = []
    for i in itertools.permutations(range(size)):
        permutation.append(i)
        
    # Initialize matrix to -1
    matrix = -np.ones((size,size))
    
    # Cycle through turns
    for turn in range(size**2):

        # Preparation for finding the permutations with most 1s and 0s
        largest_zeroes_perm = -1
        largest_ones_perm = -1
        zperms = []
        operms = []
        
        for sigma in permutation:
            no_of_zeroes = 0
            no_of_ones = 0
            no_empty = 0

            # Find number of each value in each perm
            for i in range(size):
                if matrix[i, sigma[i]] == 0:
                    no_of_zeroes += 1
                elif matrix[i, sigma[i]] == 1:
                    no_of_ones += 1
                else:
                    no_empty +=1

            # Add the perms to the list of longest perms
            if (no_empty > 0):
                if (no_of_zeroes > largest_zeroes_perm):
                    largest_zeroes_perm = no_of_zeroes
                    zperms = [sigma]
                elif (no_of_zeroes == largest_zeroes_perm):
                    zperms.append(sigma)

                if (no_of_ones > largest_ones_perm):
                    largest_ones_perm = no_of_ones
                    operms = [sigma]
                elif (no_of_ones == largest_ones_perm):
                    operms.append(sigma)

        # Both players play in a random place of a random perm
        # with the greatest number of 1s
        if turn % 2 == 0: # Alternates players
            rand_max_perm = random.choice(operms)
            free_positions = []
            for i in range(size):
                if matrix[i, rand_max_perm[i]] == -1:
                    free_positions.append(i)
            random_pos = random.choice(free_positions)
            matrix[random_pos, rand_max_perm[random_pos]] = 0
    
        else:
            rand_max_perm = random.choice(operms)
            free_positions = []
            for i in range(size):
                if matrix[i, rand_max_perm[i]] == -1:
                    free_positions.append(i)
            random_pos = random.choice(free_positions)
            matrix[random_pos, rand_max_perm[random_pos]] = 1

    # Check the product of each permutation, if all are
    # 0 it's a definite win, otherwise uncertain.
    for sigma in permutation:
        product = 1
        for i in range(size):
            product *= matrix[i, sigma[i]]
        if product == 1 :
            return "lose"
    return "win"

wins = 0
losses = 0
for game in range(100):
    if random_game(4) == "win":
        wins += 1
    else:
        losses += 1
print("Won", wins, "times\nLost", losses,"times")
```

```python
from itertools import product
import random
import numpy as np

def odd_mirror(size, first):
    # Matrix of -1s to start
    # Unncesary, could be set to zero 
    # But neater to follow
    matrix = -np.ones((size, size))

    # Keep track of the squares not yet filled 
    free_spaces = list(product(range(size), range(size)))
    free_spaces.remove(((size-1)//2, (size-1)//2))
    matrix[(size+1)//2, (size-1)//2] = first

    for turn in range(size**2 // 2):
        # Select random element not yet filled
        p1_move = random.choice(free_spaces)
        i = p1_move[0]
        j = p1_move[1]

        free_spaces.remove((i,j))
        matrix[i, j] = 1

        ## Mirror along the leading diagonal
        #free_spaces.remove((size-1-i,size-1-j))  
        #matrix[size-1-i, size-1-j] = 0

        ## Mirror the rows
        # Or columns if in the centre column.
        if j != (size-1)//2:
            free_spaces.remove((i, size-1-j))
            matrix[i, size-1-j] = 0
        else:
            free_spaces.remove((size-1-i, j))
            matrix[size-1-i, j] = 0
    
    return np.linalg.det(matrix)

# Play 100 times with matrices the size `size`
# first_move is whoever moves first
first_move = 1
for size in range(5, 52, 2):
    wins = 0
    for game in range(100):
        if odd_mirror(size, first_move) == 0:
            wins +=1
    print("Games won of 100 with size", size, ":", wins)
```
