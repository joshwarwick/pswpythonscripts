```python
import matplotlib.pyplot as plt
import numpy as np

def recursion(p):
    return 1-0.5*p
p = [1]
for i in range(24):
    p.append(recursion(p[-1]))

plt.figure(figsize=(10, 6))
plt.bar(np.arange(25)+1, p)
plt.title("Result of recursion")
plt.xlabel("Position $i$")
plt.ylabel("$p(i)$")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

lands = np.zeros(15)
lands[0] = 1
lands[1] = 0.5

for n in range(2, 15):
    for i in range(n):
        success = 0
        for combination in product([1, 2], repeat=n):
            position = 1
            for i in range(n):
                if combination[i] == 1:
                    position += 1
                else:
                    position += 2
                if position == n+1:
                    success += 1
                    break
                if position == n+2:
                    break

        lands[n] = success / 2**n

plt.figure(figsize=(10, 6))
plt.bar(np.arange(15)+1, lands)
plt.title("Result of path search")
plt.xlabel("Position $i$")
plt.ylabel("$p(i)$")
plt.show()
```
